```python
def Choose():
    choose=input("DO YOU WANT TO PLAY WITH COMPUTER AS C OR PLAYER AS P: ")
    Choose.c=choose
    if(Choose.c=="C"):
        Choose.name=input("WHAT IS YOUR NAME: ")
    elif(Choose.c=="P"):
        Choose.name1=input("WHAT IS THE NAME OF THE PLAYER 1: ")
        Choose.name2=input("WHAT IS THE NAME OF THE PLAYER 2: ")
    else:
        print("WRONG INPUT, CHOOSE BETWEEN C AND P")
        Choose()
    
    while(Choose.c!="C" and Choose.c!="P"):
        print("WRONG CHOICE!, CHOOSE AGAIN...")
        Choose()
    
    
```
